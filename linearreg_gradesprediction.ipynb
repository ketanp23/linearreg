```python
import numpy as np
import matplotlib.pyplot as plt

class SimpleLinearRegression:
    def __init__(self):
        self.slope = None
        self.intercept = None

    def fit(self, X, y):
        """
        Train the linear regression model using input features X and target y
        Formula:
        slope = (n*Σ(xy) - Σx*Σy) / (n*Σ(x²) - (Σx)²)
        intercept = (Σy - slope*Σx) / n
        """
        n = len(X)
        sum_x = np.sum(X)
        sum_y = np.sum(y)
        sum_xy = np.sum(X * y)
        sum_x_sq = np.sum(X ** 2)

        # Calculate slope and intercept
        numerator = n * sum_xy - sum_x * sum_y
        denominator = n * sum_x_sq - sum_x ** 2
        self.slope = numerator / denominator
        self.intercept = (sum_y - self.slope * sum_x) / n

    def predict(self, X):
        """Predict target values using the trained model"""
        if self.slope is None or self.intercept is None:
            raise Exception("Model not trained yet. Call fit() first.")
        return self.slope * X + self.intercept

    def plot_regression_line(self, X, y):
        """Visualize the regression line along with data points"""
        plt.scatter(X, y, color='blue', label='Actual Data')
        plt.plot(X, self.predict(X), color='red', label='Regression Line')
        plt.title('Simple Linear Regression')
        plt.xlabel('X')
        plt.ylabel('y')
        plt.legend()
        plt.grid(True)
        plt.show()

# Sample dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Create and train model
model = SimpleLinearRegression()
model.fit(X, y)

# Print learned parameters
print(f"Slope (Coefficient): {model.slope:.4f}")
print(f"Intercept: {model.intercept:.4f}")

# Make prediction
sample_x = 6
prediction = model.predict(sample_x)
print(f"Prediction for x={sample_x}: {prediction:.4f}")

# Plot results
model.plot_regression_line(X, y)
```
